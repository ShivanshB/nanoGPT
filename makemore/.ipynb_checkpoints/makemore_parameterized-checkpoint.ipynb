# Makemore Parameterized

### A parameterized implementation of makemore to allow for exploration of gradients and batching.

```python
import torch
import copy
import math
import random
import numpy as np
import torch.nn.functional as F
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
words = open('names.txt', 'r').read().splitlines()
words[:5]
```

```python
chars = sorted(set(''.join(words)))

stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0

itos = {i:s for s,i in stoi.items()}
```

```python
context_len = 3
context = [0] * context_len

X, Y = [], []
random.shuffle(words)

for w in words:    
    for char in w + '.':
        X.append(context)
        Y.append(stoi[char])  
        context = context[1:] + [stoi[char]]
        
X, Y = torch.tensor(X), torch.tensor(Y)

n1, n2 = int(len(X)*0.8), int(len(X)*0.9)
Xtr, Ytr = X[:n1], Y[:n1]
Xval, Yval = X[n1:n2], Y[n1:n2]
Xtest, Ytest = X[n2:], Y[n2:]
```

```python
# reset params to retrain
vocab_size = len(itos)
n_embed = 10
n_hidden = 200

C = torch.randn((vocab_size, n_embed))
W1 = torch.randn((n_embed * context_len, n_hidden)) * (5/3)/((n_embed * context_len)**0.5)

# ignore biases bc of batchnorm
# b1 = torch.randn(n_hidden)                          * 0.01

W2 = torch.randn((n_hidden, vocab_size))            * 0.01
b2 = torch.randn(vocab_size)                        * 0.00

bngain = torch.ones((1, n_hidden))
bnbias = torch.ones((1, n_hidden))

bnmean_running = torch.zeros((1, n_hidden))
bnstd_running = torch.ones((1, n_hidden))

parameters = [C, W1, W2, b2, bngain, bnbias]

n_params = sum(p.numel() for p in parameters)
print(f"{n_params} parameters")

for p in parameters:
    p.requires_grad = True
```

```python
# minibatch training loop
NUM_ITERS = 200000
batch_size = 32

for i in range(NUM_ITERS):
    
    # minibatch indices
    idxs = torch.randint(0, Xtr.shape[0], (batch_size,))
    Xb, Yb = Xtr[idxs], Ytr[idxs]
    
    # forward pass
    emb = C[Xb].view(-1, n_embed * context_len)
    hpreact = emb @ W1 # + b1 â€” ignore biases bc of batchnorm
    
    # running calc of stdev and mean for batchnorm
    bnmeani = hpreact.mean(0, keepdim=True)
    bnstdi = hpreact.std(0, keepdim=True)
    
    hpreact = bngain * (hpreact - bnmeani)/(bnstdi) + bnbias
    h = hpreact.tanh()
    logits = h @ W2 + b2
    
    # updating running batchnorm statistics
    with torch.no_grad():
        bnmean_running = 0.999 * bnmean_running + 0.001 * bnmeani
        bnstd_running = 0.999 * bnstd_running + 0.001 * bnstdi

    # calculating loss
    loss = F.cross_entropy(logits, Yb)
    
    # clearing old gradients
    for p in parameters:
        p.grad = None
    
    # backprop & update
    loss.backward()
    
    lr = 0.1 if i < 100000 else 0.01
    for p in parameters:
        p.data -= lr * p.grad
        
    if i % 10000 == 0:
        print(loss.item())
```

Here, we've added batch normalization to the forward and backward passes. Note that earlier (chronologically earlier, but below in the notebook), we were finding ways to make sure we don't oversaturate the activation functions and get vanishing gradients. Here, notice that we can manually normalize our outputs of each layer, so we don't need to worry about this anymore (at least mostly). Further notice that we initialize bngain and bnbias to all ones to start with a N(1,1) distribution.

Further notice that we obtained a val loss of 2.12 using batch norm, my best yet.

Lastly, note that batch normalization, by definition of batches of training points during training, stops the model from overfitting on any single sample, and thus acts as a sort of **regularizer**.

```python
# calculate batchnorm statistics over entire training dataset

with torch.no_grad():
    emb = C[Xtr].view(-1, n_embed * context_len)
    h = emb @ W1 + b1
    bnmean = h.mean(0, keepdim=True)
    bnstd = h.std(0, keepdim=True)
```

Notice that this is another step after training that needs to be completed before inference, and can become unwieldy to do as a single operation with larger datasets. Thus, we will implement above a method to calculate the mean and stdev in a running manner

We will now see how close the true bnmean and bnstd are to our running estimates.

```python
abs_diff = (bnmean - bnmean_running).abs().mean()
rel_diff = (abs_diff / bnmean.mean()).abs()
print(f"relative difference of {rel_diff * 100:.4f}%")
```

```python
# evaluate train and val loss

# @torch.no_grad() # this decorator disables gradient tracking
def split_loss(split):
    x,y = {'train': (Xtr, Ytr), 'val': (Xval, Yval), 'test': (Xtest, Ytest)}[split]
    emb = C[x] # (N, block_size, n_embd)
    embcat = emb.view(-1, n_embed * context_len) # concat into (N, block_size * n_embd)
    hpreact = embcat @ W1 + b1
    hpreact = bngain * ((hpreact - bnmean_running))/bnstd_running + bnbias
    h = hpreact.tanh() # (N, n_hidden)
    logits = h @ W2 + b2 # (N, vocab_size)
    loss = F.cross_entropy(logits, y)
    print(split, loss.item())
```

Notice here that we have what is called a "saturated" tanh (or any other activation function). Notice that most of the values of h are clustered around or exactly at -1 or +1. Recall that the gradient of tanh(x) = 1 - x^2, and thus for x = -1 or +1, this gradient --> 0 and the gradient we want to backpropogate vanishes. We will now try to fix this by making our h values closer to zero.

```python
plt.hist(hpreact.view(-1).tolist(), 50);
```

Now, notice that we don't have extreme clusters around -1 and +1 and so we should be utilizing more of our neurons. We will not try to retrain our model and see what loss we can achieve. Notice that it is not clear how to choose these scaling constants on our weight matrices. There is a generally a standard way of choosing them to avoid saturating our activation functions, which is to scale by $\alpha = \frac{1}{\sqrt{n}}$. Notice that if we want $var(z) = 1$ to avoid oversaturaton, and we assume that our node z takes in n inputs $x_i$ with $var(x_i) = 1$, then if we scale z by $\frac{1}{\sqrt{n}}$, we get $var(z \cdot \frac{1}{\sqrt{n}}) = \frac{1}{n} \cdot var(z) = \frac{1}{n} \cdot n \cdot var(x_i) = var(x_i) = 1$. Notice that here we used that $var(a \cdot x) = a^{2} \cdot var(x)$. I will demonstrate an example below.

```python
X = torch.randn((100, 20))
W = torch.randn((20, 20))
print((X@W).std(), math.sqrt(20))
```

Nowadays, many people will initialize their nn's with torch.nn.init.kaiming_normal.

Lastly, notice that since batch normalization subtracts the mean before normalizing the pre-activations, our bias terms (b1 and b2) don't affect the model and have gradients of 0. As such, they are spurious and unecessary and we can remove them. The learned bias (bnbias above) of the batchnorm layer is the new bias term that we are learning.

## Pytorch-ifying Our Model

```python
class Linear:
    
    def __init__(self, fan_in, fan_out, bias=True):
        self.weight = torch.randn((fan_in, fan_out)) / (fan_in ** 0.5)
        self.bias = torch.zeros(fan_out) if bias else None
        
    def __call__(self, x):
        self.out = x @ self.weight
        if self.bias is not None:
            self.out += self.bias
        return self.out
    
    def parameters(self):
        return [self.weight] + ([self.bias] if self.bias is not None else [])
    
class BatchNorm1d:
    
    def __init__(self, dim, eps=1e-5, momentum=0.1):
        self.eps = eps
        self.momentum = momentum
        self.training = True
        
        # params
        self.gamma = torch.ones(dim)
        self.beta = torch.zeros(dim)
        
        # buffers maintained with momentum update
        self.running_mean = torch.zeros(dim)
        self.running_var = torch.ones(dim)
        
    def __call__(self, x):
        if self.training:
            xmean = x.mean(0, keepdim=True)
            xvar = x.var(0, keepdim=True)
        else:
            xmean = self.running_mean
            xvar = self.running_var
        xhat = (x - xmean) / (torch.sqrt(xvar + self.eps))
        self.out = self.gamma * xhat + self.beta
        
        if self.training:
            with torch.no_grad():
                self.running_mean = (1 - self.momentum) * self.running_mean + self.momentum * xmean
                self.running_var = (1 - self.momentum) * self.running_var + self.momentum * xvar
        return self.out
    
    def parameters(self):
        return [self.gamma, self.beta]
    
class Tanh:
    
    def __call__(self, x):
        self.out = x.tanh()
        return self.out
    
    def parameters(self):
        return []
```

```python
# defining network
n_embed = 10
n_hidden = 100

C = torch.randn((vocab_size, n_embed))

layers = [
    Linear(n_embed * context_len, n_hidden), Tanh(),
    Linear(             n_hidden, n_hidden), Tanh(),
    Linear(             n_hidden, n_hidden), Tanh(),
    Linear(             n_hidden, n_hidden), Tanh(),
    Linear(             n_hidden, n_hidden), Tanh(),
    Linear(             n_hidden, vocab_size),
]

with torch.no_grad():
    # reduce confidence of last layer (get closer to uniform)
    layers[-1].weight *= 0.1
    
    # apply gain to the rest of the layers (keep activations from fully saturating)
    for layer in layers[:-1]:
        if isinstance(layer, Linear):
            layer.weight *= 5/3
    
# gather all parameters and calculate total number
parameters = [C] + [p for layer in layers for p in layer.parameters()]
print(f"{sum(p.nelement() for p in parameters)} parameters")

for p in parameters:
    p.requires_grad = True
```

```python
# minibatch training loop
NUM_ITERS = 200000
batch_size = 32
lossi = []
ud = []

for i in range(NUM_ITERS):
    
    # minibatch indices
    idxs = torch.randint(0, Xtr.shape[0], (batch_size,))
    Xb, Yb = Xtr[idxs], Ytr[idxs]
    
    # forward pass
    x = C[Xb].view(-1, n_embed * context_len)
    for layer in layers:
        x = layer(x)
    
    # calculating loss
    loss = F.cross_entropy(x, Yb)
    
    # allows layers to save grad
    for layer in layers:
        layer.out.retain_grad()
        
    # clearing out old gradients
    for p in parameters:
        p.grad = None
    
    # backward pass
    loss.backward()
    
    # update
    lr = 0.1 if i < 100000 else 0.01
    for p in parameters:
        p.data -= lr * p.grad
        
    # print every once in a while
    if i % 10000 == 0: 
        print(f'{i:7d}/{NUM_ITERS:7d}: {loss.item():.4f}')
    
    lossi.append(loss.log10().item())
    with torch.no_grad():
        ud.append([((lr*p.grad).std() / p.data.std()).log10().item() for p in parameters])
```

First, we claim that we want

1. gradients to be similarly distributed across layers in all depths of the model
2. not overly saturated activations for each layer

Notice that above, with only the use of Tanh (no batchnorm layers), the initializations of the gains of each weight matrix is extremely finnicky. If the gain is too higher (>> 5/3 in this case), the activations are overly saturated and the gradient distribution varies throughout the layers. Similarly, if the gain is too low (<< 5/3 in this case), then the activations go to 0 and the gradients also vary in distribution throughout the network. Notice that  the Tanh layers work to squish the values in the model and so the gain works against this to keep the values well distributed. It should be clear that if these is even a slight mismatch in the "squishing" and "expanding" forces of the activation and gain respectively, it will be magnified as we get deeper in the model and will hamper training.

We will now add batchnorm layers to our network and see how the resulting activations are much less finnicky and give us nicely distributed activations and gradients.

```python
# defining network
n_embed = 10
n_hidden = 100

C = torch.randn((vocab_size, n_embed))

layers = [
    Linear(n_embed * context_len, n_hidden, bias=False), BatchNorm1d(n_hidden), Tanh(),
    Linear(             n_hidden, n_hidden, bias=False), BatchNorm1d(n_hidden), Tanh(),
    Linear(             n_hidden, n_hidden, bias=False), BatchNorm1d(n_hidden),Tanh(),
    Linear(             n_hidden, n_hidden, bias=False), BatchNorm1d(n_hidden),Tanh(),
    Linear(             n_hidden, n_hidden, bias=False), BatchNorm1d(n_hidden),Tanh(),
    Linear(             n_hidden, vocab_size, bias=False), BatchNorm1d(vocab_size),
]

with torch.no_grad():
    # reduce confidence of last layer (get closer to uniform)
    layers[-1].gamma *= 0.1 # use gamma instead of weight bc last layer is now a batchnorm
    
    # apply gain to the rest of the layers (keep activations from fully saturating)
    for layer in layers[:-1]:
        if isinstance(layer, Linear):
            layer.weight *= 5/3
    
# gather all parameters and calculate total number
parameters = [C] + [p for layer in layers for p in layer.parameters()]
print(f"{sum(p.nelement() for p in parameters)} parameters")

for p in parameters:
    p.requires_grad = True
```

```python
# minibatch training loop
NUM_ITERS = 200000
batch_size = 32
lossi = []
ud = []

for i in range(NUM_ITERS):
    
    # minibatch indices
    idxs = torch.randint(0, Xtr.shape[0], (batch_size,))
    Xb, Yb = Xtr[idxs], Ytr[idxs]
    
    # forward pass
    x = C[Xb].view(-1, n_embed * context_len)
    for layer in layers:
        x = layer(x)
    
    # calculating loss
    loss = F.cross_entropy(x, Yb)
    
    # allows layers to save grad
    for layer in layers:
        layer.out.retain_grad()
        
    # clearing out old gradients
    for p in parameters:
        p.grad = None
    
    # backward pass
    loss.backward()
    
    # update
    lr = 0.1 if i < 100000 else 0.01
    for p in parameters:
        p.data -= lr * p.grad
        
    # print every once in a while
    if i % 10000 == 0: 
        print(f'{i:7d}/{NUM_ITERS:7d}: {loss.item():.4f}')
    
    lossi.append(loss.log10().item())
    with torch.no_grad():
        ud.append([((lr*p.grad).std() / p.data.std()).log10().item() for p in parameters])
        
    if i >= 1000:
        break
```

```python
@torch.no_grad() # this decorator disables gradient tracking
def split_loss(split):
    x,y = {'train': (Xtr, Ytr), 'val': (Xval, Yval), 'test': (Xtest, Ytest),}[split]
    emb = C[x] # (N, block_size, n_embd)
    x = emb.view(emb.shape[0], -1) # concat into (N, block_size * n_embd)
    for layer in layers:
        x = layer(x)
    loss = F.cross_entropy(x, y)
    print(split, loss.item())

# put layers into eval mode
for layer in layers:
    layer.training = False
split_loss('train')
split_loss('val')
```

```python
@torch.no_grad() # this decorator disables gradient tracking
def split_loss(split):
    x,y = {'train': (Xtr, Ytr), 'val': (Xval, Yval), 'test': (Xtest, Ytest),}[split]
    emb = C[x] # (N, block_size, n_embd)
    x = emb.view(emb.shape[0], -1) # concat into (N, block_size * n_embd)
    for layer in layers:
        x = layer(x)
    loss = F.cross_entropy(x, y)
    print(split, loss.item())

# put layers into eval mode
for layer in layers:
    layer.training = False
split_loss('train')
split_loss('val')
```

Note that we have now folded the weights of the batchnorm into the layer before it (in this case a linear layer), and as such the batchnorm is only there to stabilize training and can be removed afterward. We verify this above by obtaining the same train and validation losses before and after folding. Not only does this provide an interesting interpretation of batchnorm, but it allows us to reduce the parameters in our model in a lossless manner.
